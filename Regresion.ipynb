El objetivo de este dataset es predecir el promedio de un estudiante alcanzado en 2023.
Posee las siguientes variables:
- Variable a predecir
    - Study_year_avg: promedio alcanzado en el año 2023 por el estudiante
        - 0 - 100
- Independientes
    - Sex: sexo del estudiante
        - [0, 1] : [femenino, masculino]
    - Matric_avg: promedio matricial del estudiante
        - 0 - 100
    - Study_year: año o grado actual (2023) que cursa el estudiante
        - [0, 1, ..., 4] : [1° año, 2° año, 3° año, 4° año, postgrado]
    - Faculty: facultad del grado del estudiante en la universidad.
        - [0, 1, ..., 7] : [AgriSciences, Arts & social science, Economics & management sciences, Education, Law, Medicine & Health sciences, Military science, Science, Theology]
    - Accom_status: cuenta con alojamiento privado o con alojamiento en la universidad de Stellenbosh
        - [0, 1] : [Privado, no privado]
    - Monthly_allowance: Prestaciones mensuales que recibe el estudiante
        - [0, 1, ..., 4] : [4001 - 5000, 5001 - 6000, 6001 - 7000, 7001 - 8000, >8000]
    - Scholarship: el estudiante se encuentra en un programa de financiación
        - [0, 1] : [Sí, no]
    - Wkly_stdy_hrs: horas semanales de estudio
        - [0, 1, ..., 4] : [0, 1-3, 3-5, 5-8, >8]
    - Drinks_per_night: cuántas bebidas alcohólicas consume el estudiante cuando sale de fiesta
        - [0, 1, ..., 4] : [0, 1-3, 3-5, 5-8, >8]
    - Missed_classes: cuántas clases ha faltado el estudiante por problemas causados por el alcohol (resaca, cansancio, otras)
        - [0, 1, ..., 4] : [0, 1, 2, 3, >4]
    - Failed_modules: cuántos módulos ha reprobado el estudiante en lo que lleva de sus estudios
        - [0, 1, ..., 4] : [0, 1, 2, 3, >4]
    - Approval: los padres del estudiante aprueban que el estudiante consuma bebidas alcohólicas
        - [0, 1] : [sí, no]
    - Parent_strength: qué tan cercana es la relación del estudiante con sus padres
        - [0, 1, 2, 3] : [Muy cercana, Cercana, poco cercana, distante]

# Importación del dataset

#### Existen múltiples columnas con elementos faltantes. 
En este caso, utilizar la media poblacional para la imputación a primeras no sería buena idea, ya que no todos los datos son de tipo entero o flotante

### Antes de la imputación

Antes de ir directamente a la eliminación de registros vacíos, los valores faltantes de las columnas con tipo de dato flotante se reemplazarán con la media de esta columna. Para verificar que las medidas estadísticas del dataset no se han modificado tras esta imputación, describiremos antes y después:

```python
students.describe()
```

### Imputación

Se usará la media poblacional de la columna "Your Matric (grade 12) Average/ GPA (in %)" para reemplazar los valores faltantes.

```python
matric_grade_mean = students["Your Matric (grade 12) Average/ GPA (in %)"].mean()

students["Your Matric (grade 12) Average/ GPA (in %)"].fillna(matric_grade_mean)
```

Después de la imputación:

```python
students.isnull().sum()
```

```python
students.describe()
```

Como podemos ver, este método de imputación no ha afectado en absoluto los estadísticos del conjunto de datos.

#### Se utiliza la función 'dropna' para eliminar los registros con valores faltantes:

```python
students.dropna(inplace=True)
```

```python
students.isnull().sum()
```

### Reemplazar variables no numéricas
En este caso asignaré un conjunto de números para que el tipo categórico sea numérico en lugar de una cadena de texto

```python
students.drop(['Timestamp'], inplace=True, axis=1)

students["Your Sex?"].replace(["Female", "Male"], [0, 1], inplace=True)

students['What year were you in last year (2023) ?'].replace(
    ['2nd Year', '1st Year', '3rd Year', '4th Year', 'Postgraduate'],
    [2, 1, 3, 4, 5],
    inplace=True
)

students["Your Accommodation Status Last Year (2023)"].replace(
    ["Private accommodation/ stay with family/friends", "Non-private accommodation ie. Res"],
    [0, 1],
    inplace = True
)

students['How many classes do you miss per week due to alcohol reasons, (i.e: being hungover or too tired?)'].replace(
    ['3', '4+', '2', '1', '0'],
    [3, 4, 2, 1, 0],
    inplace=True
)
students['How many modules have you failed thus far into your studies?'].replace(
    ['0', '3', '4+', '1', '2'],
    [0, 3, 4, 1, 2],
    inplace=True
)

students["Were you on scholarship/bursary in 2023?"].replace(
    ["No", "Yes (NSFAS, etc...)"],
    [0, 1],
    inplace = True
)

students['Monthly Allowance in 2023'].replace(
[   'R 4001- R 5000', 'R 7001 - R 8000', 'R 6001 - R 7000',
    'R 5001 - R 6000', 'R 8000+'],
    [1, 4, 3, 2, 5],
    inplace=True
)

students['How often do you go out partying/socialising during the week? '].replace(
    ['Only weekends', '2', '3', '1', '4+', '0'],
    [5, 2, 3, 1, 4, 0],
    inplace=True
)

students["On a night out, how many alcoholic drinks do you consume?"].replace(
    ['8+', '3-5', '5-8', '1-3', '0'],
    [4, 2, 3, 1, 0],
    inplace=True
)

students["Additional amount of studying (in hrs) per week"].replace(
    ['8+', '3-5', '1-3', '5-8', '0'],
    [5, 3, 2, 4, 1],
    inplace=True
)

students["What faculty does your degree fall under?"].replace(
    ['Arts & Social Sciences', 'Economic & Management Sciences',
       'AgriSciences', 'Engineering', 'Science', 'Law',
       'Medicine and Health Services', 'Education'],
    [0, 1, 2, 3, 4, 5, 6, 7],
    inplace = True
)

students["Are you currently in a romantic relationship?"].replace(
    ["No", "Yes"],
    [0, 1],
    inplace = True
)

students["Do your parents approve alcohol consumption?"].replace(
    ["No", "Yes"],
    [0, 1],
    inplace = True
)

students["How strong is your relationship with your parent/s?"].replace(
    ['Very close', 'Fair', 'Close', 'Distant'],
    [3, 1, 2, 0],
    inplace=True
)
```

### Estandarización de los datos

Ya que los datos de tipo flotante son calificaciones con rango [0, 100], únicamente dividiré los datos de la columna entre 100:

```python
students['Your Matric (grade 12) Average/ GPA (in %)'] = students['Your Matric (grade 12) Average/ GPA (in %)'] / 100
students['Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)'] = students['Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)'] / 100
```

```python
students.dtypes
```

Como podemos observar, nos quedamos únicamente con tipos de dato numéricos, por lo que podemos pasar a la siguiente fase de este notebook.

# Modelos de regresión utilizados

- Regresión lineal
- SVM (máquina de soporte vectorial)
- Decision tree
- Random forest

## Separación de los conjuntos de entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split

X = students.drop(['Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)'], axis=1)
y = students['Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)'].values
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.75)
```

## Regresión lineal

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_train, y_train)
y_pred_lr = lr.predict(X_test)
```

```python
plt.scatter(range(len(y_test)), y_test, color='blue', label='Valores Reales')
plt.scatter(range(len(y_test)), y_pred_lr, color='red', label='Predicciones', marker='x')
plt.xlabel('Índice de Muestra')
plt.ylabel('Valor')
plt.title('Predicciones vs Valores Reales')
plt.legend()
plt.show()
```

### Métricas de rendimiento para la Regresión Lineal

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred_lr_1 = np.array(y_pred_lr).reshape(1, -1)
y_test_lr_1 = np.array(y_test).reshape(1, -1)

print(f"Mean absolut error: {mean_absolute_error(y_test,y_pred_lr)}")
print(f"Mean square error: {mean_squared_error(y_test,y_pred_lr)}")
print(f"R^2 score:", r2_score(y_test, y_pred_lr))
```

## SVM

```python
from sklearn.svm import SVR
import matplotlib.pyplot as plt

svr = SVR(kernel='rbf')

svr.fit(X_train, y_train)
```

#### Métricas de rendimiento para la SVM

```python
y_pred_svm = svr.predict(X_test)

y_test_2 = y_test.reshape(1, -1)
y_pred_svm_2 = y_pred_svm.reshape(1, -1)
```

```python
plt.scatter(range(len(y_test)), y_test, color='blue', label='Valores Reales')
plt.scatter(range(len(y_test)), y_pred_svm, color='red', label='Predicciones', marker='x')
plt.xlabel('Índice de Muestra')
plt.ylabel('Valor')
plt.title('Predicciones vs Valores Reales')
plt.legend()
plt.show()
```

### Métricas de rendimiento para la SVM

```python
print(f"Mean Absolut error: {mean_absolute_error(y_test,y_pred_svm)}")
print(f"Mean squared error: {mean_squared_error(y_test,y_pred_svm)}")
print(f"R^2 score:", r2_score(y_test, y_pred_svm))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor() 

rfr = rfr.fit(X_train,y_train)  
y_pred_rfr = rfr.predict(X_test)
```

```python
y_test_rfr_2 = y_test.reshape(1, -1)
y_pred_rfr_2 = y_pred_rfr.reshape(1, -1)
```

```python
plt.scatter(range(len(y_test)), y_test, color='blue', label='Valores Reales')
plt.scatter(range(len(y_test)), y_pred_rfr, color='red', label='Predicciones', marker='x')
plt.xlabel('Índice de Muestra')
plt.ylabel('Valor')
plt.title('Predicciones vs Valores Reales')
plt.legend()
plt.show()
```

### Métricas de rendimiento para random forest

```python
print(f"Mean Absolut error: {mean_absolute_error(y_test,y_pred_rfr)}")
print(f"Mean squared error: {mean_squared_error(y_test,y_pred_rfr)}")
print(f"R^2 score:", r2_score(y_test, y_pred_rfr))
```

## Decision tree

```python
from sklearn.tree import DecisionTreeRegressor

d_tree = DecisionTreeRegressor() 
d_tree_train = d_tree.fit(X_train,y_train)
y_pred_d_tree = d_tree_train.predict(X_test)
```

### Métricas de rendimiento para el decision tree

```python
print(f"Mean Absolut error: {mean_absolute_error(y_test,y_pred_d_tree)}")
print(f"Mean squared error: {mean_squared_error(y_test,y_pred_d_tree)}")
print(f"R^2 score:", r2_score(y_test, y_pred_d_tree))
```
